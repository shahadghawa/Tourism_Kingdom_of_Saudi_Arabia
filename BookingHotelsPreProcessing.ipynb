### Preprossing

```python
#Droup unnecessary columns 
df=df.dropna(subset=['Rate'])
df=df.dropna(subset=['Score'])
df=df.drop(['Unnamed: 0.1','Unnamed: 0'], axis=1)
df
```

```python
#Fill null value with coordinates
df.loc[df['Title'] == "شقة نيرفانا ريزدنس Nirvana Residence Apt", 'latlon'] = '24.833817422758095,46.6050683124813'
```

```python
#Change the column type 
df['Checkin_day']=df['Check_in'].dt.day_name()
df['Checkout_day']=df['Check_Out'].dt.day_name()
```

```python
#Change the type again to extract the day and month to new column 
df['Check_in']=df['Check_in'].astype(str)
df['Check_Out']=df['Check_Out'].astype(str)
```

```python
#Spilt the column for check in and rename it  
split=df['Check_in'].str.split('-',expand=True)
df['Year']=split[0]
df['Month']=split[1]
df['Day']=split[2]
```

```python
#Spilt the columne for check out and rename it 
split=df['Check_Out'].str.split('-',expand=True)
df['Year']=split[0]
df['Month']=split[1]
df['Day']=split[2]
```

```python
#Change the column type to date time again 
df['Check_in']= pd.to_datetime(df['Check_in'])
df['Check_Out']= pd.to_datetime(df['Check_Out'])
```

```python
#Change the column type for the month  
df['Check_Month']=df['Check_in'].dt.month_name()
```

```python
#Spilt the lotlon in tow columne column 
split=df['latlon'].str.split(',',expand=True)
#Rename it 
df['lon']=split[0]
df['lat']=split[1]
```

```python
#Split the Room type column
split=df['RoomType'].str.split(' ',expand=True)
df['Type']=split[0]
```

```python
df.loc[df['Type'] == ' شقة', 'TYPE'] = 'شقة'
df.loc[df['Type'] == ' استوديو', 'TYPE'] = 'Studio'
```

```python
#Catogry for the Type
df.loc[df['Type'] == "طماط", 'Type'] = "جناح"
df.loc[df['Type'] == "غرف", 'Type'] = "غرفة"
df.loc[df['Type'] == "قياسية", 'Type'] = "غرفة"
df.loc[df['Type'] == "قياسية،", 'Type'] = "غرفة"
df.loc[df['Type'] == "وحدة", 'Type'] = "غرفة"
df.loc[df['Type'] == "تنفيذية،", 'Type'] = "غرفة"
df.loc[df['Type'] == "وحدة", 'Type'] = "غرفة"
df.loc[df['Type'] == "Standard", 'Type'] = "غرفة"
df.loc[df['Type'] == "خيار", 'Type'] = "غرفة"
df.loc[df['Type'] == "كينغ", 'Type'] = "جناح"
df.loc[df['Type'] == "الجناح", 'Type'] = "جناح"
df.loc[df['Type'] == "شاليه", 'Type'] = "شاليه"
df.loc[df['Type'] == "سوبيريور", 'Type'] = "شاليه"
df.loc[df['Type'] == "فيلا", 'Type'] = "منزل"
df.loc[df['Type'] == "بيت", 'Type'] = "منزل"
df.loc[df['Type'] == "استديو", 'Type'] = "استوديو"
df.loc[df['Type'] == "ستوديو", 'Type'] = "استوديو"
# *****************************************************
df.loc[df['Type'] == "استديو", 'Type'] = "استوديو"
df.loc[df['Type'] == "ستوديو", 'Type'] = "استوديو"
```

```python
df.loc[df['Type'] == 'استوديو', 'Type'] = 'Studio'
df.loc[df['Type'] == "شاليه", 'Type'] = 'Chalet'
df.loc[df['Type'] == "شقة", 'Type'] = 'Apartment'
df.loc[df['Type'] == "شاليه", 'Type'] = 'Chalet'
df.loc[df['Type'] == "منزل", 'Type'] = 'House'
df.loc[df['Type'] == "جناح", 'Type'] = 'Suite'
df.loc[df['Type'] == "غرفة", 'Type'] = 'Room'
```

```python
#Drop the column  
df=df.drop(['RoomType'], axis=1)
df.head()
```

```python
#Split the column
split=df['Airport'].str.split(',',expand=True)
df['Airportlon']=split[0]
df['Airportlat']=split[1]
```

```python
#Change the type for the coordinate column  
df["Airportlon"]=df["Airportlon"].astype(float)       
df["Airportlat"]=df["Airportlat"].astype(float)  
df["lat"]=df["lat"].astype(float)  
df["lon"]=df["lon"].astype(float)  
```

```python
import h3
#calculate the distance between the hotel and the airport 
df['Distance'] = df.apply(lambda row: h3.point_dist((row['Airportlon'], row['Airportlat']), (row['lon'], row['lat'])), axis=1)
```
