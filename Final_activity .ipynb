### Import Libraries

```python
# Data Processiong library
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read the CSV file



```python
#read the dataset
activity=pd.read_csv('ThingsToDoRiyadh.csv')
activity
```

### Droup unneeded columne 

```python
#Droup the column not need 
activity=activity.drop(['Unnamed: 0'], axis=1)
activity=activity.drop(['level_0'], axis=1)
activity=activity.drop(['index'], axis=1)
activity=activity.drop(['URLS'], axis=1)
activity
```

```python
#Cheak the null value
activity.isna().sum()
```

```python
#Cheak the summary of the dataset
activity.info()
```

```python
activity.describe()
```

```python
#Make a copy of the dataset 
activity1 = activity.copy()
activity1
```

### The Over all columne 

```python
#split the OverAll column 
test= activity1['OverAll'].str.split('s.', expand=True)
#Rename the split column 
activity1['Rating by Reviews']=test[0]
#Droup the column after split 
activity1=activity1.drop(['OverAll'], axis=1)
#Clean the column
activity1['Rating by Reviews'] = activity1['Rating by Reviews'].str.replace(r'of 5 bubble', '')
activity1['Rating by Reviews'] = activity1['Rating by Reviews'].str.replace(r"\['", '')
#clean the column 
activity1['Review'] = activity1['Review'].str.replace(r"review", '')
activity1
```

### Rank columne 

```python
#Clean column 
activity1['Rank'] = activity1['Rank'].str.replace(r"\['#", '')
activity1
```

### The Time columne 

```python
#split the column 
test2= activity1['Time'].str.split('- ', expand=True)
#rename the new column 
activity1['Time of Openinig']=test2[0]
activity1['Time of Closing']=test2[1]
#clean the new column
activity1['Time of Openinig'] = activity1['Time of Openinig'].str.replace(r"\['", '')
activity1['Time of Closing'] = activity1['Time of Closing'].str.replace(r"\']", '')
#drop the un need column 
activity1=activity1.drop(['Time'], axis=1)
#Deeling with null value 
activity1['Time of Openinig']=activity1['Time of Openinig'].fillna('Pray time')
activity1['Time of Closing']=activity1['Time of Closing'].fillna('Pray time')
activity1
```

### The title columne 

```python
#clean the column 
activity1['Title'] = activity1['Title'].str.replace(r"\']", '')
activity1['Title'] = activity1['Title'].str.replace(r"\['", '')
activity1
```

### The activety columne 

```python
activity1['Type of activety']=activity1['Unnamed: 8']
```

```python
#clean the column 
activity1['Type of activety'] = activity1['Type of activety'].str.replace(r"\']", '')
activity1['Type of activety'] = activity1['Type of activety'].str.replace(r"\['", '')
activity1['Type of activety'] = activity1['Type of activety'].str.replace(r"88", '')
activity1['Type of activety'] = activity1['Type of activety'].str.replace(r"30", '')
activity1['Type of activety'] = activity1['Type of activety'].str.replace(r"84", '')
activity1['Type of activety'] = activity1['Type of activety'].str.replace(r"27", '')
activity1['Type of activety'] = activity1['Type of activety'].str.replace(r"8", '')
activity1['Type of activety'] = activity1['Type of activety'].str.replace(r"10", '')
activity1['Type of activety'] = activity1['Type of activety'].str.replace(r"5", '')
activity1['Type of activety'] = activity1['Type of activety'].str.replace(r"29", '')
activity1['Type of activety'] = activity1['Type of activety'].str.replace(r"46", '')
activity1['Type of activety'] = activity1['Type of activety'].str.replace(r" in Riyadh", '')
activity1=activity1.drop(['Unnamed: 8'], axis=1)
activity1['Type of activety']
```

```python
activity1['Type of activety'].value_counts()
```

# EDA

```python
activity1['Rank'].fillna(0,inplace=True)
activity1['Rank']=activity1['Rank'].astype(int)
activity1['Rating by Reviews']=activity1['Rating by Reviews'].astype(float)
activity1['Rating by Reviews'].fillna(0,inplace=True)
```

```python
activitysort=activity1[0:5]
activitysort
```

```python
activitysort['long']=[24.7313,24.6553,24.6314,24.7682,24.6772]
activitysort['lat']=[46.6700,46.7092, 46.7136,46.7149,46.7788]
```

```python
import folium
map_osm=folium.Map(location=[24.7313,46.6700],zoom_start=50)
```

```python
for i in range(0,len(activitysort)):
    html=f"""
        <h3> {activitysort.iloc[i]['Title']}</h3>
        <p>Info:</p>
        <ul>
            <li>Rank:{activitysort.iloc[i]['Rank']} </li>
            <li>Rating:{activitysort.iloc[i]['Rating by Reviews']} out of 5 </li>
            <li>Reviews:{activitysort.iloc[i]['Review']} </li>
             <li>Activity:{activitysort.iloc[i]['Type of activety']} </li>
        </ul>
        </p>
        """
    iframe = folium.IFrame(html=html, width=200, height=200)
    popup = folium.Popup(iframe, max_width=2650)
    folium.Marker(
        location=[activitysort.iloc[i]['long'], activitysort.iloc[i]['lat']],
        popup=popup,
       icon=folium.Icon(color="orange",icon="users", prefix='fa')
    ).add_to(map_osm)
```

```python
from branca.element import Figure
fig2=Figure(width=550,height=550)
fig2.add_child(map_osm)
folium.TileLayer('Stamen Terrain').add_to(map_osm)
folium.TileLayer('Stamen Toner').add_to(map_osm)
folium.TileLayer('Stamen Water Color').add_to(map_osm)
folium.TileLayer('cartodbpositron').add_to(map_osm)
folium.TileLayer('cartodbdark_matter').add_to(map_osm)
folium.LayerControl().add_to(map_osm)
```
